```python
def find_best_correlation(df, target_column):
    best_corr_pair = None
    best_corr_value = 0  # Initialize with a value that cannot be a correlation coefficient

    for column in df.columns:
        if column != target_column:
            try:
                correlation_value = df[target_column].corr(df[column])
                if abs(correlation_value) > abs(best_corr_value):
                    best_corr_value = correlation_value
                    best_corr_pair = (target_column, column)
            except Exception as e:
                print(
                    f"Error calculating correlation between {target_column} and {column}: {str(e)}")

    correlation_matrix = df.corr()

    return best_corr_pair, best_corr_value, correlation_matrix
```
